# Preprocessamento

```python
y_train = train['NU_NOTA_MT'].fillna(0)
train.drop(columns='NU_NOTA_MT', inplace=True)

X_train = train[test.columns]
X_test = test.copy()
```

```python
X_train.dtypes
```

## Categóricas

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, 
from sklearn.pipeline import Pipeline
```

```python
feat_cat = [*X_train.select_dtypes(include='object').columns]
```

```python
feat_cat
```

Primeiramente, vou executar o *encoder* nos dados completos para obter todas as categorias:

```python
full = (X_train[feat_cat].append(X_test[feat_cat])).copy()
temp_pipe = Pipeline([
    ('imp', SimpleImputer(strategy='most_frequent')),
    ('enc', OneHotEncoder(drop='if_binary', sparse=False))
])
temp_fit = temp_pipe.fit(full)

all_categories = temp_fit.named_steps['enc'].categories_
del full
```

Agora, definindo o pipeline para as features categóricas:

```python
cat_tr = Pipeline([
    ('cat_imp', SimpleImputer(strategy='most_frequent')),
    ('cat_enc', OneHotEncoder(categories=all_categories, drop='if_binary', sparse=False))
])
```

# Numéricas

Duas estratégias de imputação: uma para as notas, outra para as demais variáveis numéricas.

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
```

```python
feat_num = [*X_train.select_dtypes(exclude='object').columns]
```

Vamos descartas algumas features:

- `CO_UF_RESIDENCIA`: redundante por causa da feature `SG_UF_RESIDENCIA`

```python
feat_num.remove('CO_UF_RESIDENCIA') # redundante
```

Separando as features de notas das outras features numéricas:

```python
feat_nota = [ft for ft in feat_num if 'NU_NOTA' in ft]
feat_num = [ft for ft in feat_num if 'NU_NOTA' not in ft]
```

```python
feat_nota, feat_num
```

Definindo o pipeline para as variáveis numéricas:

```python
num_tr = Pipeline([
    ('num_imp', SimpleImputer(strategy='mean')),
    ('num_sc', StandardScaler())
])

nota_tr = Pipeline([
    ('nota_imp', SimpleImputer(fill_value=0)),
    ('nota_poly', PolynomialFeatures(degree=2, include_bias=False))
])
```

## Consolidado

```python
from sklearn.compose import ColumnTransformer
```

```python
data_tr = ColumnTransformer([
    ('cat_tr', cat_tr, feat_cat),
    ('num_tr', num_tr, feat_num),
    ('nota_tr', nota_tr, feat_nota),
], remainder='drop')
```

# Modelos

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import ElasticNet, Lasso, LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor, BaggingRegressor, GradientBoostingRegressor, RandomForestRegressor
from sklearn import metrics
```

Vamos testar os modelos usando uma validação cruzada de 4 *folds* no conjunto de testes de validação:

```python
kfold = KFold(n_splits=4)
```

```python
models = [
    LinearRegression(),
    ElasticNet(tol=0.00001),
    Lasso(tol=0.00001),
    KNeighborsRegressor(),
    DecisionTreeRegressor(),
    AdaBoostRegressor(),
    BaggingRegressor(),
    GradientBoostingRegressor(),
    RandomForestRegressor()
]
```

```python
models_names = ['linreg','elastic', 'lasso', 'knn', 'decision', 'adaboost', 'bagging', 'gradient', 'random_forest']
```

```python
scores = {}
```

```python
X_transf = data_tr.fit_transform(X_train)

for index, model in enumerate(models):
    print(f'Treinando modelo {models_names[index]}')
    scores[models_names[index]] = cross_val_score(model, X_transf, y_train, cv=kfold)
```

```python
scores
```

# Modelo final

```python
from sklearn.model_selection import GridSearchCV
from sklearn.feature_selection import RFECV
from sklearn.pipeline import make_pipeline
```

```python
mod = GradientBoostingRegressor(loss='huber', max_depth=10)
```

```python
mod.get_params()
```

## Seleção de features

```python
rfe = RFECV(LinearRegression(), cv=kfold, step=.1, n_jobs=-1)
```

## Hiperparâmetros

```python
param_grid = {
    'learning_rate': [0.01, 0.1, 0.2],
    'min_samples_leaf': [1,2],
    'n_estimators': [50, 100],
    'subsample': [0.8, 1]
}
```

```python
grid = GridSearchCV(mod, param_grid=param_grid, cv=3, refit=True, verbose=2, n_jobs=-1)
```

```python
pipe = Pipeline([
    ('data_processing', data_tr),
    ('feature_sel', rfe)
])

grid.fit(pipe.fit_transform(X_train, y_train), y_train)
```

## Consolidado

```python
pipe = Pipeline([
    ('data_processing', data_tr),
    #('feature_sel', rfe),
    ('model', grid.best_estimator_)
])
```

```python
pipe.fit(X_train, y_train)
```

# Avaliação do modelo

```python
preds_train = pipe.predict(X_train)
```

```python
metrics.r2_score(y_train, preds_train)
```

```python
metrics.median_absolute_error(y_train, preds_train)
```

```python
plt.scatter(x=y_train, y=preds_train, alpha=0.3)
plt.ylabel('Notas previstas')
plt.xlabel('Notas reais')
```

# Final

```python
preds = pipe.predict(X_test)
```

Exportando os resultados:

```python
out = pd.DataFrame({'NU_INSCRICAO': test.index, 'NU_NOTA_MT': preds})
out.head()
```
